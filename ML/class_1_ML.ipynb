#  Scaling And Encoding

```python
import pandas as pd

df = pd.DataFrame(
{'Age': [23, 20, 30],
'experience':[ 3, 1, 7],
'salary':[3000, 3500, 5000]}
)

df
```

```python
from sklearn.preprocessing import MinMaxScaler


min_max_scaler = MinMaxScaler()
min_max_scaler.fit(df[['Age', 'experience', 'salary']])

df[['AgeScaled', 'experience_scaled', 'salary_scaled']] = min_max_scaler.transform(df[['Age', 'experience', 'salary']])
df
```

```python
from sklearn.preprocessing import StandardScaler


standard_scaler = StandardScaler()
standard_scaler.fit(df[['Age', 'experience', 'salary']])

df[['AgeScaled', 'experience_scaled', 'salary_scaled']] = standard_scaler.transform(df[['Age', 'experience', 'salary']])
df
```

```python
from sklearn.preprocessing import RobustScaler


robust_scaler = RobustScaler()
robust_scaler.fit(df[['Age', 'experience', 'salary']])

df[['AgeScaled', 'experience_scaled', 'salary_scaled']] = robust_scaler.transform(df[['Age', 'experience', 'salary']])
df
```

```python
from sklearn.preprocessing import MaxAbsScaler


MaxAbs_scaler = MaxAbsScaler()
MaxAbs_scaler.fit(df[['Age', 'experience', 'salary']])

df[['AgeScaled', 'experience_scaled', 'salary_scaled']] = MaxAbs_scaler.transform(df[['Age', 'experience', 'salary']])
df
```

```python
df = pd.DataFrame({
    'size':['small', 'large', 'medium', 'medium', 'small'],
'brand':['addidas', 'nike', 'nike', 'addidas', 'addidas'], 
'warranty': [ 1, 2, 2, 1.5, 1],
'color': ['red', 'blue', 'red', 'blue', 'green'],
'price': [100, 150, 200, 250, 300]
})

df
```

```python
from sklearn.preprocessing import OrdinalEncoder


size_order = ['small', 'medium', 'large']

ordinal_encoder = OrdinalEncoder(categories=[size_order])

ordinal_encoder.fit(df[['size']])

df['size_transformed'] = ordinal_encoder.transform(df[['size']])
df['size_transformed'] = df['size_transformed'].astype(int)

df
```
