```python
def adder(*args):
    product = 1
    for number in args:
        product = product * number
    return product

result = adder(5,6)
print(result)
```

```python
def calculator(*args, **kwargs):
    if kwargs["ops"] == "product":
        product = 1
        for number in args:
            product = product * number
        return product
    elif kwargs["ops"] == "add":
        return sum(args)
    else:
        return f"invalid operation : {kwargs["ops"]}"
result = calculator(5, 6,ops="add")
print(result)
```

```python
def calculate_density(mass, volume):
    try:
         den = mass / volume
    except ZeroDivisionError:
        return "volume cannot be zero"
    except TypeError:
        return "invalid data type"
    except:
        return "invalid input"
    return den
```

```python
def calculate_bmi(weight, height):
    try:
        bmi = weight / (height**2)
    except ZeroDivisionError:
        return "height cannot be zero"
    except TypeError:
        return "invalid data type"
    except:
        return " invalid input"
    return bmi
```

```python
bmi_value = calculate_bmi(100, 0)
bmi_values = calculate_bmi("a", 180)
bmi_right = calculate_bmi(150, 1.8)
print(bmi_value)
print(bmi_values)
print(bmi_right)
```

```python
def calculate_density(mass, volume):
    if type(mass) not in (int, float):
        raise TypeError("mass has to be a number")
    if mass <= 0:
        raise ValueError("mass has to be a positive value")
    if type(volume) not in (int, float):
        raise TypeError("volume has to be a number")
    if volume <= 0:
        raise ValueError("volume has to be a positive number")
    density = mass / volume
    return density
```

```python
den = calculate_density(10, 10)
print(den)
```

```python
class Point:
    def __init__(self, x_cord, y_cord):
        self.x_cord = x_cord
        self.y_cord = y_cord

    @property
    def Magnitude(self):
        Magnitude = (self.y_cord ** 2 + self.y_cord ** 2) ** 0.5
        return Magnitude
    def __repr__(self):
        return f"P({self.x_cord}, {self.y_cord})"
    
    def __len__(self):
        return round(self.Magnitude)
    def __add__(self, other):
        new_x = self.x_cord + other.x_cord
        new_y = self.y_cord + other.y_cord
        new_point = Point(new_x, new_y)
        return new_point
    
    def __lt__(self, other):
        return self.Magnitude < other.Magnitude
```

```python
class Account:
    def __init__(self, account_holder, account_number, amount=0):
        self.__amount = amount
        self.account_holder = account_holder
        self.account_number = account_number

    def get_balance(self):
        return self.__amount  
    def deposit(self, deposit_amount):
        if deposit_amount < 10:
            raise ValueError("deposit amount cannot be less than 10")
        self.__amount = self.__amount + deposit_amount  
    def withdraw(self, withdraw_amount):
        if withdraw_amount > 99999:
            raise ValueError("withdraw amount cannot exceed 99999. ")  
        self.__amount = self.__amount - withdraw_amount 
        if withdraw_amount > self.__amount:
            raise ValueError ("withdrawn amount cannot exceed the bank balance. ")
    def transfer(self, other, amount):
        self.withdraw(amount)
        other.deposit(amount)
```

```python
Account_1 = Account("ram", "1234")
account_2 = Account("ram", "12332", 10000)
```

#  ABSTRACTION

```python
from abc import ABC, abstractmethod

class shape(ABC):
    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def perimeter(self):
        pass
```

```python
class rectangle(shape):
    def __init__(self, length, breadth):
        self.length = length
        self.breadth = breadth

    def area(self):
        return self.length * self.breadth

    def perimeter(self):
        return 2 * (self.length + self.breadth)
```

```python
r1 = rectangle(2,5)
print(r1.area())
print(r1.perimeter())
```
