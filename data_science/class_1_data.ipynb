```python
def calculate_grade(row):
    if not row['grade'] is not None:
        return 'grade'
    if row['attendance'] > 80:
        if row['score']  > 80:
            return 'A+'
        elif row['score'] > 70:
            return 'A'
        elif row['score'] > 60:
            return 'B'
        else :
            return 'C'
    else :
        if row['score'] > 80:
            return 'A'
        elif row['score'] > 70:
            return 'B'
        elif row['score'] > 60:
            return 'C'
        else :
            return 'D'
```
